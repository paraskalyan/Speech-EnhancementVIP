```python
import numpy as np
import scipy
from scipy.fftpack import fft
import librosa
import matplotlib.pyplot as plt
import scipy.io.wavfile as wavfile
import scipy as sp
import scipy.fftpack as fftpk
import numpy as np
from matplotlib import pyplot as plt
import IPython.display as ipd
```

```python
infile = "a60.mp3"
print ('load wav', infile)
w, sr = librosa.load( infile, sr=None, mono=True) # keep native sr (sampling rate) and trans into mono
ff = abs(fft(w))
freqs = np.fft.fftfreq(len(ff), (1.0/sr))
plt.plot(freqs[range(len(ff)//2)], ff[range(len(ff)//2)])                                                          
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
# X = fft(w)
# plt.plot(abs(X))
# # s= librosa.stft(w) 
# # ss= np.abs(s)         # get magnitude
# plt.plot(abs(s))
# plt.show()
# Short-time Fourier transform

# angle= np.angle(s)    # get phase
# b=np.exp(1.0j* angle)
# dur = librosa.get_duration(y=w, sr=sr)
# N = sr*dur
# print(N)
# use this phase information when Inverse Transform
```

```python
# output = scipy.signal.wiener(ff)
from scipy.signal import wiener
filt = wiener(ff,mysize = 50)
filt
```

```python
# import numpy as np
out = np.fft.ifft(filt)
out2 = out.real
print(out2)
```

```python

import soundfile as sf
sf.write("out.wav", out2, 48000, 'PCM_24')
ipd.Audio("out.wav", rate=sr)
plt.plot(out2)
plt.show
```
